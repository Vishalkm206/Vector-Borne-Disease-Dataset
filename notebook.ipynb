```python
import pandas as pd
import numpy as np

train_dataset = pd.read_csv("train.csv",header=0)
train_dataset.head()
```

I think ID have nothing to deal with what disease a person has
do our x will be from column 1 to end-1
and our y will be end column

```python
x = train_dataset.iloc[:,1:-1]
y = train_dataset.iloc[:,-1]
print(x.shape)
print(y.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
model.fit(x,y)
```

```python
train_pred = model.predict(x)
accuracy = accuracy_score(y, train_pred)
print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
df = pd.read_csv("test.csv",header=0)
test = df.iloc[:,1:]
test.head()
```

```python
test["prognosis"] = model.predict(test)
```

```python
submission = pd.DataFrame([df["id"],test["prognosis"]])
submission = submission.T
submission.to_csv("RealSubmission.csv",index=False)
```
